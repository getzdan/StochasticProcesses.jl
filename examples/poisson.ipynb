```julia
push!(LOAD_PATH, "$(pwd())/../src"); using PyPlot, StatsBase, Distributions, StochasticProcesses;
```

```julia
let p=PoissonProcess(1000)
    plt[:hist](sim(p, linspace(0, 1, 10000), 10000), bins=50, normed=true, range=(800,1200));
    (x->plot(x, pdf(distribution(p, 1.), x), color="red"))(800:1200);
end;
```

```julia
let t=linspace(0., 1., 1000), p=PoissonProcess(500)
    plot(t, cumsim(p, t, 1000), color=(0,0,0,.02))
end;
```
