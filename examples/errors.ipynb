## Strong convergence 

```julia
let p=GeometricBrownianMotion(.1, .2, 100.), f=(t,b)->p.y0 * exp(t * (p.mu - p.sigma^2 / 2) + p.sigma * b)
    errors = Vector{Float64}()
    for s in 1:20
        t=linspace(0,1, 2^s)
        result = sim(p, t, 1000)
        push!(errors, sqrt(mean((result .- f(t[end], result.b)) .^2)))
    end
    semilogy(errors)
end;
```
