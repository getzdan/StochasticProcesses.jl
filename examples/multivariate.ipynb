## Multivariate Brownian Motion

```julia
let t=linspace(0, 1, 1000), d = cumsim(BrownianMotion([1, 2]), t)
    plot(t, d[:,1])
    plot(t, d[:,2])
end;
```

```julia
let t=linspace(0, 1, 1000), d = sim(BrownianMotion([1, 2]), t, 10000)
    scatter(x=d[1,:], y=d[2,:], alpha=.1, marker=".")
end;
```
